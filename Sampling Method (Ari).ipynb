# Type of Sampling


## 1. Random Sampling

```python
#Choose 200 resp (random)
df.sample(200)
```

## 2. Systematic Sampling

```python
#Choose data drom all data (0-1802, multiple of 10)
df.iloc[0:1802:10]
```

## 3. Stratisfied Sampling (ex: Gender) 

```python
#Create var df... of each gender
df_male=df[df["Gender"]=="M"]
df_female=df[df["Gender"]=="F"]
```

```python
#Choose random 100 Female Resp and 100 Male Resp
random_gender=[df_male.sample(100), df_female.sample(100)]
pd.concat(random_gender)
```
